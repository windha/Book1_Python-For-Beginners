```python
import numpy as np
import matplotlib.pyplot as plt

# 定义二维向量
vector_a = np.array([3, 4])    # 向量 a = (3, 4)
vector_b = np.array([1, 2])    # 向量 b = (1, 2)
vector_c = np.array([-2, 3])   # 向量 c = (-2, 3)

print(f"向量 a: {vector_a}")
print(f"向量 b: {vector_b}")
print(f"向量 c: {vector_c}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体支持
plt.rcParams['font.sans-serif'] = ['SimHei', 'Arial Unicode MS', 'DejaVu Sans']
plt.rcParams['axes.unicode_minus'] = False  # 正确显示负号

# 创建向量
vector_a = np.array([3, 4])
vector_b = np.array([1, 2])
vector_sum = vector_a + vector_b  # 向量相加

# 创建图形
plt.figure(figsize=(10,8))

# 绘制向量 a (从原点到 (3, 4))
plt.quiver(0, 0, vector_a[0], vector_a[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.005, label='向量 a = [3, 4]')

# 绘制向量 b (从原点到 (1, 2))
plt.quiver(0, 0, vector_b[0], vector_b[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.005, label='向量 b = [1, 2]')

# 绘制向量和 a + b
plt.quiver(0, 0, vector_sum[0], vector_sum[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='green', width=0.005, label='向量 a + b = [4, 6]')

# 绘制平行四边形（用于展示向量加法）
plt.quiver(vector_a[0], vector_a[1], vector_b[0], vector_b[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='red', width=0.003, alpha=0.5, linestyle='--')

plt.quiver(vector_b[0], vector_b[1], vector_a[0], vector_a[1], 
           angles='xy', scale_units='xy', scale=1, 
           color='blue', width=0.003, alpha=0.5, linestyle='--')

# 设置图形属性
plt.xlim(-1, 6)
plt.ylim(-1, 6)
plt.grid(True, alpha=0.3)
plt.axhline(y=0, color='k', linewidth=0.5)
plt.axvline(x=0, color='k', linewidth=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('二维向量可视化')
plt.legend()
plt.axis('equal')

# 添加坐标标注
plt.annotate(f'a({vector_a[0]}, {vector_a[1]})', 
             xy=(vector_a[0], vector_a[1]), xytext=(vector_a[0]+0.3, vector_a[1]+0.3))
plt.annotate(f'b({vector_b[0]}, {vector_b[1]})', 
             xy=(vector_b[0], vector_b[1]), xytext=(vector_b[0]+0.3, vector_b[1]+0.3))
plt.annotate(f'a+b({vector_sum[0]}, {vector_sum[1]})', 
             xy=(vector_sum[0], vector_sum[1]), xytext=(vector_sum[0]+0.3, vector_sum[1]+0.3))

plt.show()

print(f"向量 a: {vector_a}")
print(f"向量 b: {vector_b}")
print(f"向量和: {vector_sum}")
```
